```python
import pandas as pd
import os
```

## Load Pitch Tracks

```python
# '../example/pitch/103_203_birthday_ground_clueGiver.csv'
# '../example/pitch/108_208_birthday_board_clueGiver.csv'

# Load the CSV file into a DataFrame
file_path = '../example/pitch/108_208_birthday_board_clueGiver.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)
df = df.drop(columns=['ID'])
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
# Check for exact match with your value

# 103_203_birthday_ground_clueGiver
#search_value = 14897.478447276939 

# 108_208_birthday_board_clueGiver
search_value = 8472.584846882472 
```

```python
# Check if the value exists within a small tolerance (for floating point comparison)
tolerance = 1e-10
close_matches = df[abs(df['time_ms'] - search_value) < tolerance]
print(f"Number of close matches found: {len(close_matches)}")
if len(close_matches) > 0:
    print("Close matches:")
    print(close_matches[['time_ms']].head())
```

```python
target_row = close_matches.index.tolist()[0]
start_index = max(0, target_row - 5)  # Ensure we don't go below index 0
end_index = min(len(df), target_row + 6)  # End index is exclusive, so add 6 to include 5 rows after

# Extract the rows
rows_around_target = df.iloc[start_index:end_index]

# Display the rows
print(rows_around_target)
```

## Clean Up Pitch Tracks

```python
# Create a copy of the original DataFrame
new_df = df.copy()

# Set all entries in f0 column to NaN after the target row
new_df.loc[target_row + 1:, 'f0'] = None

target_row = close_matches.index.tolist()[0]
start_index = max(0, target_row - 5)  # Ensure we don't go below index 0
end_index = min(len(df), target_row + 6)  # End index is exclusive, so add 6 to include 5 rows after

# Extract the rows
rows_around_target = new_df.iloc[start_index:end_index]

# Display the rows
print(rows_around_target)
```

```python
# Extract directory and filename
directory = os.path.dirname(file_path)
filename = os.path.basename(file_path)

# Create the cleaned directory path
cleaned_directory = os.path.join(directory, "cleaned")

# Create the cleaned directory if it doesn't exist
os.makedirs(cleaned_directory, exist_ok=True)

# Combine cleaned directory and filename
output_file = os.path.join(cleaned_directory, filename)

# Export the dataframe
new_df.to_csv(output_file, index=False)
```
