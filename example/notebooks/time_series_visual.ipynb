```python
import pandas as pd
import scipy
from scipy.signal import correlate
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mark_non_nan(time, values, color='red', alpha=0.1):
    valid = ~np.isnan(values)

    changes = np.diff(valid.astype(int))
    starts = np.where(changes == 1)[0] + 1
    ends = np.where(changes == -1)[0] + 1

    if valid[0]:
        starts = np.r_[0, starts]
        ends = np.r_[ends, len(valid)]

    for s, e in zip(starts, ends):
        plt.axvspan(time[s], time[e-1], color=color, alpha=alpha)
```

### 103_203

```python
file_103 = '../merged/103_203.csv'  # Replace with your actual file path
df_103 = pd.read_csv(file_103)

time_103 = df_103['t'].to_numpy()
eye_103 = df_103['eyebrows_interp'].to_numpy()
pitch_raw_103 = df_103['pitch'].to_numpy()
pitch_103 = scipy.stats.zscore(pitch_raw_103, nan_policy='omit')
```

```python
plt.figure(figsize=(12, 4))
plt.plot(time_103, eye_103, label="Eyebrows")
plt.plot(time_103, pitch_103, label="Pitch")
plt.xlabel("Time")
plt.xlim(0, time_103.max())
mark_non_nan(time_103, pitch_103)
#plt.ylabel("Value")
plt.title("Birthday 103_203 : Clue giver on ground")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Get boolean mask for non-NaN values
mask = ~np.isnan(pitch_103)
pitch_filtered_103 = pitch_103[mask]
eye_filtered_103 = eye_103[mask]
```

```python
plt.figure(figsize=(12, 4))
plt.plot(eye_filtered_103, label="Eyebrows")
plt.plot(pitch_filtered_103, label="Pitch")
plt.xlabel("Time")
plt.title("Birthday 103_203 : Clue giver on ground (NaN filtered)")
plt.legend()
plt.grid(True)
plt.show()
```
